# 7.What is the average state of charge at the start and end of charging?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming you've already calculated these values
avg_soc_start = df['State of Charge (Start %)'].mean()
avg_soc_end = df['State of Charge (End %)'].mean()

# Data for the plot
soc_data = pd.DataFrame({
    'SOC Stage': ['Start SOC (%)', 'End SOC (%)'],
    'Average SOC (%)': [avg_soc_start, avg_soc_end]
})

# Set the style of seaborn plot
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='SOC Stage', y='Average SOC (%)', data=soc_data, palette='Blues')

# Add labels and title
plt.title('Average State of Charge (Start vs End)', fontsize=14)
plt.ylabel('State of Charge (%)')
plt.ylim(0, 100)  # Limit the y-axis to the range of SOC percentages

# Show the plot
plt.show()

```

# Conclusion 
 The average State of Charge (SOC) at the start of charging sessions is approximately 48.67%, and the SOC at the end of charging is around 79.86%

# 10.What is the distribution of charging sessions across different days of the week?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of charging sessions for each day of the week
charging_sessions_by_day = df['Day of Week'].value_counts()

# Plotting the distribution of charging sessions
plt.figure(figsize=(8, 6))
sns.barplot(x=charging_sessions_by_day.index, y=charging_sessions_by_day.values, palette='Purples')

# Add labels and title
plt.title('Distribution of Charging Sessions by Day of the Week', fontsize=14)
plt.xlabel('Day of the Week')
plt.ylabel('Number of Charging Sessions')

# Show the plot
plt.show()

```

# Conclusion
The majority of charging sessions occur on weekdays, with the lowest number observed on Sundays, indicating higher charging demand during workdays.

# 11.What is the average distance driven since the last charge for different user types?

```python
# avg_distance_by_user_type = ev_data.groupby('User Type')['Distance Driven (since last charge) (km)'].mean()
# avg_distance_by_user_type
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average distance driven since the last charge for each user type
avg_distance_by_user_type = df.groupby('User Type')['Distance Driven (since last charge) (km)'].mean().reset_index()

# Plotting the average distance driven for different user types
plt.figure(figsize=(8, 6))
sns.barplot(x='User Type', y='Distance Driven (since last charge) (km)', data=avg_distance_by_user_type, palette='Blues')

# Add labels and title
plt.title('Average Distance Driven Since Last Charge by User Type', fontsize=14)
plt.xlabel('User Type')
plt.ylabel('Distance Driven (km)')

# Show the plot
plt.show()
```

# Conclusion
Long-distance travelers drive the highest average distance since their last charge, while casual drivers cover the least distance, reflecting distinct charging and driving patterns across user types.

# 12. How does the average energy consumed differ across vehicle models?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate the average energy consumed for each vehicle model
avg_energy_by_vehicle_model = df.groupby('Vehicle Model')['Energy Consumed (kWh)'].mean().reset_index()

# Sorting the models by average energy consumption for better visualization
avg_energy_by_vehicle_model = avg_energy_by_vehicle_model.sort_values(by='Energy Consumed (kWh)', ascending=False)

# Plotting the average energy consumed for different vehicle models
plt.figure(figsize=(10, 6))
sns.barplot(x='Energy Consumed (kWh)', y='Vehicle Model', data=avg_energy_by_vehicle_model, palette='Greens')

# Add labels and title
plt.title('Average Energy Consumed by Vehicle Model', fontsize=14)
plt.xlabel('Energy Consumed (kWh)')
plt.ylabel('Vehicle Model')

# Show the plot
plt.show()

```

# Conclusion
The average energy consumption varies significantly across vehicle models, with certain models consuming noticeably higher energy per charging session, likely due to differences in battery capacity and usage patterns.

# 15. What is the total energy consumption by each vehicle model?

```python
# total_energy_by_vehicle_model = df.groupby('Vehicle Model')['Energy Consumed (kWh)'].sum()
# total_energy_by_vehicle_model
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the total energy consumption for each vehicle model
total_energy_by_vehicle_model = df.groupby('Vehicle Model')['Energy Consumed (kWh)'].sum().reset_index()

# Sorting the models by total energy consumption for better visualization
total_energy_by_vehicle_model = total_energy_by_vehicle_model.sort_values(by='Energy Consumed (kWh)', ascending=False)

# Plotting the total energy consumed for different vehicle models
plt.figure(figsize=(10, 6))
sns.barplot(x='Energy Consumed (kWh)', y='Vehicle Model', data=total_energy_by_vehicle_model, palette='Blues')

# Add labels and title
plt.title('Total Energy Consumption by Vehicle Model', fontsize=14)
plt.xlabel('Total Energy Consumed (kWh)')
plt.ylabel('Vehicle Model')

# Show the plot
plt.show()
```

# conclusion
 The total energy consumption varies significantly across vehicle models, with some models consuming more energy overall due to larger battery sizes or more frequent charging sessions.
